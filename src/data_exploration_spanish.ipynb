#### Spanish Kaggle Dataset

```python
df2_test = pd.read_csv('../data/Spanish_Kaggle/test.csv')
df2_test
```

### Merge

```python
df1m_train = df1_train[['Text', 'Category']]
df1m_train.columns = ['text', 'label']
df1m_train['label'].replace(to_replace='Fake', value=1, inplace=True)
df1m_train['label'].replace(to_replace='True', value=0, inplace=True)
df1m_train.index.name = 'id'
df1m_train
```

```python
df1m_test = df1_test[['Text', 'Category']]
df1m_test.columns = ['text', 'label']
df1m_test['label'].replace(to_replace='Fake', value=1, inplace=True)
df1m_test['label'].replace(to_replace='True', value=0, inplace=True)
df1m_test
```

```python
df2m_train = df2_train[['text', 'label']]
df2m_train.index.name = 'id'
df2m_train
```

```python
merge = df1m_train.append(df1m_test, ignore_index = True)
merge.index.name = 'id'
merge = merge.append(df2m_train, ignore_index = True)
merge.index.name = 'id'
merge
```

#### Save (export) dataset

```python
merge.to_csv('../data/Merged/spanish_dataset.csv', index = False)
```
