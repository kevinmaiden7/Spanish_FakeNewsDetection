```python
from googletrans import Translator
import pandas as pd
```

```python
dataset = pd.read_csv('../data/Merged/spanish_dataset.csv')
dataset
```

```python
translator = Translator()
dataset['text_traslated'] = dataset['text'].apply(translator.translate, src='es', dest='en').apply(getattr, args=('text',))
dataset
```

```python
dataset.to_csv('../data/Merged/spanish_traslated_dataset.csv', index = False)
```
