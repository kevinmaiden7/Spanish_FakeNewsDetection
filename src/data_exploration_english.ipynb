```python
import numpy as np
import pandas as pd
```

#### English 0

```python
df0_fake = pd.read_csv('../data/English_0/Fake.csv')
df0_fake
```

```python
df0_true = pd.read_csv('../data/English_0/True.csv')
df0_true
```

#### English 1

```python
df1 = pd.read_csv('../data/English_1/news_dataset.csv')
df1
```

```python
df1.label.describe()
```

#### Processing

```python
del(df0_fake['title'])
del(df0_fake['subject'])
del(df0_fake['date'])
df0_fake['label'] = 1
df0_fake
```

```python
del(df0_true['title'])
del(df0_true['subject'])
del(df0_true['date'])
df0_true['label'] = 0
df0_true
```

```python
del(df1['Unnamed: 0'])
del(df1['title'])
df1['label'].replace(to_replace='FAKE', value=1, inplace=True)
df1['label'].replace(to_replace='REAL', value=0, inplace=True)
df1
```

```python
# Export df1 as small balanced English dataset
df1.to_csv('../data/English_1/small_english_dataset.csv', index = False)
```

#### Merge

```python
merge = df0_fake.append(df0_true, ignore_index = True)
merge.index.name = 'id'
merge = merge.append(df1, ignore_index = True)
merge.index.name = 'id'
merge
```

#### Save (export) dataset

```python
merge.to_csv('../data/Merged/english_dataset.csv', index = False)
```
